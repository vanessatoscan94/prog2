```python
df = pd.json_normalize(data)

# Convert numerical values in the column
df['Aadorf.resident.Aadorf.Wohnbevölkerung'] = df['Aadorf.resident.Aadorf.Wohnbevölkerung'].str.replace("'", "").astype(int)

# Create bar chart
df.plot(x='Aadorf.city_name', y='Aadorf.resident.Aadorf.Wohnbevölkerung', kind='bar')
plt.xlabel('Stadt')
plt.ylabel('Wohnbevölkerung')
plt.title('Wohnbevölkerung in Aadorf')
plt.show()
```
